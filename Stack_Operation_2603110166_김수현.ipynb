```python
stack = []
capacity = 6

def isFull() :
    if len(stack) == capacity :
        return True
    else :
        return False

def isEmpty() :
    if len(stack) == 0 :
        return True
    else :
        return False


def StackStatus() :
    print(f'[Stack Status] {stack}')

def Push(data) :
    if isFull() :
        print('the stack is Full!')
    else :
        stack.append(data)
        
        StackStatus()

def Pop() :
    if isEmpty() :
        print('the stack is Empty!')
    else :
        data = stack.pop()
        print(f'[Pop] 가져온 데이터 : {data}')

        StackStatus()

def Peek() :
    if isEmpty() :
        print('the stack is Empty!')
    else :
        data = stack[ len(stack) - 1 ]
        print(f'>> top 데이터 확인 : {data}')

        StackStatus()

print(f'[[정수형 스택 연산 실습 (용량 : {capacity})]]')

while True :

    print('===================================')
    print(' 1. Push  2. Pop  3. Peek  0. Exit')
    print('===================================')

    menu = int(input('>> 연산 메뉴 선택:'))

    if menu == 1 :
        data = int(input('[Push] 추가할 데이터 입력 :'))
        Push(data)

    elif menu == 2 :
        Pop()

    elif menu == 3 :
        Peek()
        
    elif menu == 0 :
        print('정수형 스택 연산 실습 종료')
        break

    else :
        print('Error : menu 범위 밖')
        continue
```
