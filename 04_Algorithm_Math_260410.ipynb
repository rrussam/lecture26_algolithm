```python
def is_prime_number(N) : 
    if N == 2 : return True

    if N <= 1 or N % 2 == 0 : 
        return False
        
    for i in range(3, int(N / 2), 2) :
        if N % i == 0 :
            return False
            
    return True


#num = int(input())
#print(is_prime_number(num))

for i in range(1, 16) :
    print(f'{i} isPrime? : {is_prime_number(i)}')
```
