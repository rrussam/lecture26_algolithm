```python
lst = []


while True :
    
    print('===================================')
    print(' 1. Push  2. Pop  3. Peek  0. Exit')
    print('===================================')

    menu = int(input('>> 연산 메뉴 선택:'))
    
    if menu == 0 :
        break
        
    elif menu == 1 :
        data = int(input('[Push] 추가할 데이터 입력 :'))
        lst.append(data)
        
    elif menu == 2 :
        data = lst.pop()
        print(f'[Pop] 가져온 데이터 : {data}')

    elif menu == 3 :
        if len(lst) - 1 > 0 :
            data = lst[len(lst) - 1]
        else :
            print('>> stack empty.')
            continue
            
        print(f'>> top 데이터 확인 : {data}')
        continue

    print(f'[Stack] {lst}')


print('정수형 스택 종료')
```

```python
import sys
lst = []
q = int(input())
for i in range(q) :
    prs= sys.stdin.readline().split()
    cmd = prs[0]
    ln = len(lst)
    if cmd == 'push' :
        lst.append(prs[1])
    if cmd == 'pop' :
        if ln == 0 : 
            print(-1)
        else :
            print(lst.pop())
    if cmd == 'size' :
        print(ln)
    if cmd == 'empty' :
        if ln == 0 :
            print(1)
        else :
            print(0)
    if cmd == 'top' :
        if ln == 0 :
            print(-1)
        else :
            print(lst[ln - 1])
```
