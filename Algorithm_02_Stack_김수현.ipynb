```python
lst = []


while True :
    
    print('===================================')
    print(' 1. Push  2. Pop  3. Peek  0. Exit')
    print('===================================')

    menu = int(input('>> 연산 메뉴 선택:'))
    
    if menu == 0 :
        break
        
    elif menu == 1 :
        data = int(input('[Push] 추가할 데이터 입력 :'))
        lst.append(data)
        
    elif menu == 2 :
        data = lst.pop()
        print(f'[Pop] 가져온 데이터 : {data}')

    elif menu == 3 :
        if len(lst) - 1 > 0 :
            data = lst[len(lst) - 1]
        else :
            print('>> stack empty.')
            continue
            
        print(f'>> top 데이터 확인 : {data}')
        continue

    print(f'[Stack] {lst}')


print('정수형 스택 종료')
```

```python
import sys
lst = []
q = int(input())
for i in range(q) :
    prs= sys.stdin.readline().split()
    cmd = prs[0]
    ln = len(lst)
    if cmd == 'push' :
        lst.append(prs[1])
    if cmd == 'pop' :
        if ln == 0 : 
            print(-1)
        else :
            print(lst.pop())
    if cmd == 'size' :
        print(ln)
    if cmd == 'empty' :
        if ln == 0 :
            print(1)
        else :
            print(0)
    if cmd == 'top' :
        if ln == 0 :
            print(-1)
        else :
            print(lst[ln - 1])
```

```python
# 단어 순서 뒤집기

N = int(input())
prs = [_ for _ in range(N)]

for i in range(N) :
    prs[i] = input()

for i in range(N) :
    word = prs[i].split()
    stack = [_ for _ in range(len(word))]
    print(f'Case #{i+1}: ', end = '')
    for j in range(len(stack)) :
        stack[j] = word.pop()
        print(stack[j], end = '')
        if j != len(stack) - 1 :
            print(end = ' ')
    print()
```

```python
# 괄호 문자열 판별

N = int(input())
prs = [_ for _ in range(N)]

for i in range(N) :
    prs[i] = input()

for i in range(N) :
    sym = list(prs[i])
    cnt = 0
    isVps = True
    
    for j in range(len(sym)) :
        if sym[j] == '(' :
            cnt += 1
        elif sym[j] == ')' :
            cnt -= 1
            if cnt < 0 :
                isVps = False

    if cnt != 0 :
        isVps = False
        
    if sym[0] == ')' or sym[len(sym) - 1] == '(' :
        isVps = False

    if isVps == True :
        print('YES')
    else :
        print('NO')
```

```python
prs = []
index = 0

while True :
    data = input()
    prs.append(data)
    if prs[index] == '.' :
        break
    index += 1

for i in range(len(prs) - 1) :
    stack = []
    sym = list(prs[i])
    isVps = True
    top = ''
    
    for j in range(len(sym)) :
        if sym[j] == '[' or sym[j] == '(' :
            stack.append(sym[j])
            top = sym[j]

        if sym[j] == ']' :
            if top == '(' or len(stack) == 0 :
                isVps = False
                break
            elif top == '[' and len(stack) != 0 :
                stack.pop()
                if len(stack) > 0 :
                    top = stack[len(stack) - 1]
                else :
                    top = ''

        if sym[j] == ')' :
            if top == '[' or len(stack) == 0 :
                isVps = False
                break
            elif top == '(' and len(stack) != 0 :
                stack.pop()
                if len(stack) > 0 :
                    top = stack[len(stack) - 1]
                else :
                    top = ''

    if len(stack) != 0 :
        isVps = False
            

    if isVps :
        print('yes')
    else :
        print('no')
    print(stack)
```
