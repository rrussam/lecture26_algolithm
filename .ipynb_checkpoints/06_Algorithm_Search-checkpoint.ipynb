```python
def binary_search_r(A, key, low, high) :
    if (low > high) :
        return -1 # 찾는 키가 A에 없는 경우 탐색할 항목이 더 없음

    middle = (low + high) // 2 # 중앙값 찾기

    if (key == A[middle]) : # 중앙값과 비교하여 같으면 그대로 반환
        return middle
    elif (key < A[middle]) : # 중앙값보다 작으면 탐색 영역을 왼쪽으로 설정
        return binary_search_r(A, key, low, middle - 1)
    else : # 중앙값보다 크면 탐색 영역을 오른쪽으로 설정
        return binary_search_r(A, key, middle + 1, high)
```

```python
def binary_search(A, key) :
    
    # index = 0
    
    # for data in A :
    #     if data == key :
    #         return index
    #     index += 1

    # return -1

    low = 0
    high = len(A) - 1

    while low <= high :
        #중앙값을 찾기
        middle = (low + high) // 2

        #중앙값과 같으면 해당 인덱스 그대로 반환
        if key == A[middle] :
            return middle
        #작으면, 왼쪽 영역 탐색 high값 변경
        elif key < A[middle] :
            high = middle - 1
        #크면, 오른쪽 영역 탐색 low값 변경
        else :
            low = middle + 1
    return -1
```

```python
data = [2,6,11,13,18,20,22,27,29,30,34,38,41,42,45,47]

num = int(input())
index = binary_search_r(data, num, 0, len(data))

if index == -1 :
    print('null value')
    
else :
    print(f'>>> num {num}은 index {index}에 있습니다.')

index = binary_search(data, num)

if index == -1 :
    print('null value')
    
else :
    print(f'>>> num {num}은 index {index}에 있습니다.')
```

```python
test_data = [34,2,47,15,48,0]
answer = [10,0,15,-1,-1,-1]

for i in range(6) :
    if binary_search(data, test_data[i]) == answer[i] :
        print('Pass : ', test_data[i])
    else :
        print('Fail : ', test_data[i])
```
