```python
# 리스트와 반복 구조
# 주어진 데이터가 어디에 저장되어 있는지 출력

data_list = [2, 42, 56, 13, 78]
target = 56
index = 0

while index <= 5 and data_list[index] != target:
    index += 1
    
print(f'>> index : {index}')
```

```python
# 리스트에서 가장 큰 값 찾기

data_list = [2, 42, 56, 13, 78]

max = data_list[0]
max_index = 0

for i in range(5) :
    if max < data_list[i] :
        max = data_list[i]
        max_index = i

print(f'>> max : {max}, max index : {max_index}')
```

```python
# 리스트에서 가장 큰 값 찾기

data_list = [2, 42, 56, 13, 78]

max = data_list[0]
index = 0
max_index = 0

for data in data_list :
    if max < data :
        max = data
        max_index = index
    index += 1

print(f'>> max : {max}, max index : {max_index}')
```

```python
# 리스트에서 가장 작은 값 찾기

data_list = [2, 42, 56, 13, 78]

min = data_list[0]
index = 0
min_index = 0

for data in data_list :
    if min > data :
        min = data
        min_index = index
    index += 1

print(f'>> min : {min}, min index : {min_index}')
```

```python
# 리스트에 있는 수들 중 입력값보다 큰 수가 몇개 있는지

data_list = [2, 42, 56, 13, 78]

num = int(input('숫자 입력 :'))
cnt = 0

for data in data_list :
    if data > num:
        cnt += 1

print(f'입력된 수보다 큰 숫자는 {cnt}개입니다')
```

```python
# 리스트에 있는 수 전부 더해서 출력

data_list = [2, 42, 56, 13, 78]
sum = 0

for data in data_list :
    sum += data

print(f'리스트 안의 수 총합은 {sum}')
```

```python
# 리스트에 있는 수 역순으로 저장하여 출력

data_list = [2, 42, 56, 13, 78]

temp = 0

for i in range(5) :
    temp = data_list[i]
    data_list.insert(5, temp)

for i in range(5) :
    data_list.pop(0)

print(data_list)
```
