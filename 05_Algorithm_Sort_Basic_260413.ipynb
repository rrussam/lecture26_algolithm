```python
def print_step(step, A) :
    print(f'{step:2} 단계 : {A}')
```

```python
# 선택 정렬
# 정렬 안된 부분(start 이후)에서 가장 작은 수를 찾아서
# start에 있는 수와 스왑
# start += 1

def select(A, start) :
    min = A[start]
    min_index = start

    for i in range(start+1, len(A)) :
        if A[i] < min :
            min = A[i]
            min_index = i
    return min_index

def SelectionSort(data_list) :
    for start in range(len(data_list) - 1) :
        min_index = select(data_list, start)
        data_list[start], data_list[min_index] = data_list[min_index], data_list[start]
        print_step(start + 1, data_list)

A = [23, 5, 44, 1, 6]
print('선택 정렬')
print('original data :', A)

SelectionSort(A)
print('sorted data :', A)
```

```python
# 삽입 정렬
# 정렬 안된 부눈의 원소를 하나씩, 정렬된 부분중 적절한 위치에 삽입하여 정렬

def InsSort(A) :
    for cur_index in range(1, len(A)) :
        # 해당 데이터의 위치를 찾아서 삽입
        key = A[cur_index]
        i = cur_index - 1
        while i >= 0 and A[i] > key :
            A[i + 1] = A[i]
            i -= 1
        # 작거나 같은 데이터의 오른쪽에 저장
        A[i + 1] = key
        print_step(cur_index, A)

A = [23, 5, 44, 1, 6]
print('삽입 정렬')
print('original data :', A)

InsSort(A)
print('sorted data :', A)
```
